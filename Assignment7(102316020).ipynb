Q1. You are a data analyst at an e-commerce company. Your task is to analyze sales data,
clean and process it, perform staƟsƟcal analysis, and generate visual insights to help the
business make informed decisions.


Part I: Randomized Sales Data GeneraƟon (NumPy)

```python
import numpy as np
import pandas as pd

# Use your roll number here (e.g., 2102456)
roll_number = 102316020
np.random.seed(roll_number)
```

```python

sales_data = np.random.randint(1000, 5001, size=(12, 4))
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df = pd.DataFrame(sales_data, index=months, columns=columns)
df.head()
```

Part II: Data ManipulaƟon and Analysis (Pandas)

```python
print("First 5 rows:")
print(df.head())

print("\nSummary Statistics:")
print(df.describe())
```

```python
total_per_category = df.sum()
total_per_month = df.sum(axis=1)

print("\nTotal per category:\n", total_per_category)
print("\nTotal per month:\n", total_per_month)
```

```python
growth = df.pct_change().mean()
print("\nAverage Sales Growth by Category:\n", growth)
```

```python
df['Total Sales'] = df.sum(axis=1)
df['Growth Rate (%)'] = df['Total Sales'].pct_change() * 100
df
```

```python
if roll_number % 2 == 0:
    df['Electronics (Discounted)'] = df['Electronics'] * 0.9
else:
    df['Clothing (Discounted)'] = df['Clothing'] * 0.85
```

Part III: Visualizations (Matplotlib & Seaborn)



```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df[columns].plot(kind='line', figsize=(12, 6), marker='o')
plt.title('Monthly Sales Trend by Category')
plt.ylabel('Units Sold')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[columns])
plt.title('Sales Distribution by Category')
plt.ylabel('Sales')
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percenƟle of flaƩened array, for each column, for each
row.
iii. Mean, Median and Standard DeviaƟon of flaƩened array, of each column, and
each row

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])
```

i. Element-wise Absolute Value

```python
abs_array = np.abs(array)
print("Absolute Values:\n", abs_array)
```

ii. Percentiles

```python

flat = array.flatten()
p_flat = np.percentile(flat, [25, 50, 75])
print("Percentiles of flattened array:", p_flat)


p_col = np.percentile(array, [25, 50, 75], axis=0)
print("Percentiles per column:\n", p_col)


p_row = np.percentile(array, [25, 50, 75], axis=1)
print("Percentiles per row:\n", p_row)
```

iii. Mean, Median, Std Deviation

```python
# Flattened
print("Mean (flat):", np.mean(flat))
print("Median (flat):", np.median(flat))
print("Std Dev (flat):", np.std(flat))

# Per column
print("Mean per column:", np.mean(array, axis=0))
print("Median per column:", np.median(array, axis=0))
print("Std Dev per column:", np.std(array, axis=0))

# Per row
print("Mean per row:", np.mean(array, axis=1))
print("Median per row:", np.median(array, axis=1))
print("Std Dev per row:", np.std(array, axis=1))
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
i. Find floor, ceiling and truncated value, rounded values

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print("Original:", a)
print("Floor:", np.floor(a))
print("Ceil:", np.ceil(a))
print("Truncated:", np.trunc(a))
print("Rounded:", np.round(a))  
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
lst = [10, 20, 30, 40, 50]
i, j = 1, 3  

temp = lst[i]
lst[i] = lst[j]
lst[j] = temp

print("Swapped list:", lst)
```

Q.5 Write a Python program to swap two elements in a set by converƟng it to a list first.

```python
s = {1, 2, 3, 4, 5}
lst = list(s)


lst[0], lst[-1] = lst[-1], lst[0]


swapped_set = set(lst)

print("Original set:", s)
print("Swapped set:", swapped_set)
```
